```python
import json
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- STYLE CONFIGURATION ---
plt.rcParams["font.family"] = "serif"  # Force Serif font
plt.rcParams["font.weight"] = "normal"  # Force Normal weight

# 1. Setup
LOG_DIR = "./logs"
files = glob.glob(os.path.join(LOG_DIR, "*.json"))
data = []

# 2. Parse Logs
for f in files:
    filename = os.path.basename(f)

    # Identify Strategy
    if "EWC" in filename:
        strategy = "EWC"
    elif "REMIND" in filename:
        strategy = "REMIND"
    elif "Naive" in filename:
        strategy = "Naive"
    else:
        continue

    # Identify Mode
    mode = "Pre-Trained" if "pretrained" in filename else "Scratch"

    # Load Data
    with open(f, "r") as json_file:
        log_data = json.load(json_file)

    if "results" in log_data and len(log_data["results"]) > 0:
        final_accs = log_data["results"][-1]["accuracies"]
        avg_acc = (sum(final_accs) / len(final_accs)) * 100

        # Create a display label
        label = f"{strategy}\n({mode})"
        data.append(
            {"Label": label, "Accuracy": avg_acc, "Strategy": strategy, "Mode": mode}
        )

# 3. Create DataFrame & Sort
df = pd.DataFrame(data)
# Sort Logic: Naive -> EWC -> REMIND(Scratch) -> REMIND(Pretrained)
sorter = ["Naive", "EWC", "REMIND"]
df["StratRank"] = df["Strategy"].astype("category").cat.set_categories(sorter).cat.codes
df.sort_values(["StratRank", "Accuracy"], inplace=True)

# 4. Generate Plot with Gradient
plt.figure(figsize=(10, 6))
sns.set_style("white")  # Clean background

# Gradient Palette (mako is a nice teal/blue gradient)
ax = sns.barplot(
    data=df, x="Label", y="Accuracy", palette="mako", edgecolor="black", linewidth=0.5
)

# Add Labels (Normal Font)
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(
            p.get_x() + p.get_width() / 2.0,
            height + 1,
            f"{height:.1f}%",
            ha="center",
            fontsize=12,
            fontname="serif",
        )

# Formatting
plt.title(
    "Continual Learning Benchmarks: Impact of Pre-training",
    fontsize=16,
    fontname="serif",
    pad=15,
)
plt.ylabel("Average Accuracy (%)", fontsize=12, fontname="serif")
plt.xlabel("")
plt.ylim(0, 100)
sns.despine()  # Removes top/right borders for a cleaner look

plt.tight_layout()
plt.savefig("benchmark_all_methods.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter DataFrame from previous step to just REMIND
df_remind = df[df["Strategy"] == "REMIND"].copy()
df_remind.sort_values("Accuracy", inplace=True)  # Low (Scratch) -> High (Pre-trained)

# --- PLOT ---
plt.figure(figsize=(6, 6))

# Use a specific 2-color gradient (Red to Green-ish)
custom_gradient = ["#d65f5f", "#488f31"]  # Soft Red, Soft Green

ax = sns.barplot(
    data=df_remind,
    x="Label",
    y="Accuracy",
    palette=custom_gradient,
    edgecolor="black",
    linewidth=0.5,
)

# Add Labels
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(
            p.get_x() + p.get_width() / 2.0,
            height + 1,
            f"{height:.1f}%",
            ha="center",
            fontsize=14,
            fontname="serif",
        )

# Formatting
plt.title("Effect of Feature Maturity on REMIND", fontsize=14, fontname="serif", pad=15)
plt.ylabel("Average Accuracy (%)", fontsize=12, fontname="serif")
plt.xlabel("")
plt.ylim(0, 100)
sns.despine()

plt.tight_layout()
plt.savefig("remind_ab_test.png", dpi=300)
plt.show()
```
